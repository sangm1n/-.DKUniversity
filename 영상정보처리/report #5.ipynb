##문제 1:
a. 그레이스케일로 읽어온 영상을 flatten 시키고, shape 을 이용하여 확인하기

```
import cv2
import matplotlib.pyplot as plt

# grayscale로 이미지 읽어오기
org_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
print('original image shape :', org_img.shape)
ftn_img = org_img.flatten()
print('flatten image shape :', ftn_img.shape)
```

b. flatten 된 이미지를 3차원 shape 형태의 이미지로 바꾸고, 형태가 바뀐 이미지를 그레이스케일 이미지로 디스플레이하기.

```
re_ftn_img = ftn_img.reshape(-1, 512, 512)
print('3-d flatten image shape :', re_ftn_img.shape)

plt.imshow(re_ftn_img[0], cmap="gray")
plt.show()
```

##문제 2
위의 읽은 원래 이미지를 이용하여, y 축 방향으로 삼등분하여, 맨 윗쪽부터 각 하부영역 이미지를 Blue, Green, Red 성분만을 최대치로 변경한 이미지를 컬러이미지 형태로 디스프레이 하기. 

```
org_img = cv2.imread(image_path)
# change BGR to RGB
org_img = org_img[:, :, ::-1]

base = int(org_img.shape[0] / 3)
start, end = 0, base-1

# B, G, R 채널 순서로 seq 리스트에 저장
seq = [2, 1, 0]
for i in range(3):
  if seq[i] == 0:
    org_img[start:, :, seq[i]] = 255
  else:
    org_img[start:end, :, seq[i]] = 255

  start += base
  end += base

plt.imshow(org_img, cmap="gray")
plt.show()
```
