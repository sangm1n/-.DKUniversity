```
import cv2

img = cv2.imread('../Dongkeun-OpenCV-ImgData/logo.png')
print('image type : {}'.format(type(img)))
print('image size : {}'.format(img.size))
print('image dtype : {}'.format(img.dtype))
```

다음의 셀에 읽은 이미지에서 좌표 y = 100, x = 50 에 있는 화소의 각 색요소 값을 한번에 읽어 출력하시오. 또한 각 색요소의 값을 별도로 가져오는 예를 작성하시오.

```
(b, g, r) = img[100, 50]
print('(b, g, r) :', b, g, r)

blue = img[100, 50, 0]
green = img[100, 50, 1]
red = img[100, 50, 2]
print('blue : {}, green : {}, red : {}'.format(blue, green, red))
```

다음의 셀에 읽은 이미지에서 좌표 y = 100, x = 50 의 화소를 순수 녹색(green) 으로 세팅하는 코드를 작성하고 해당 분야를 ROI 를 이용해 확대해서 출력하시오. 

```
import matplotlib.pyplot as plt

img[100, 50] = (0, 255, 0)
ROI = img[90:110, 40:60]
plt.imshow(ROI)
plt.show()
```

### What is Numpy ?
Numpy는 <code>Numerical Python</code>의 줄임말로 벡터 및 행렬 연산에 있어 편리한 기능을 제공한다.

```
# module import
import numpy as np
```

0-Dimensional array

```
arr = np.array(7)
print('0-D array :', arr)
print('0-D array shape :', arr.shape)
```

1-Dimensional array

```
arr = np.array([4.1, 3, -8.5, 0.9])
print('1-D array :', arr)
print('1-D array shape :', arr.shape)
```

2-Dimensional array

```
arr = np.array([[100, 50, 75],
                [45, 78, 10],
                [32, 45, 97],
                [7, 56, 88]])
print('2-D array :', arr)
print('2-D array shape :', arr.shape)
```

3-Dimensional array

```
arr = np.array([[[5.2, 4.8],
                 [8.1, 9.7]],
                [[0.3, 0.8],
                 [8.2, 6.6]]])
print('3-D array :', arr)
print('3-D array shape :', arr.shape)
```

```
grade = np.array([93, 95, 84, 100, 99, 87])

# slicing
print('slice 95~99 :', grade[1:5])

# reverse
print('reverse :', grade[::-1])

# 차원 변경 (1차원 -> 2차원)
print('1-D shape :', grade.shape)
new_grade = grade.reshape(3, 2)
print('2-D shape :', new_grade.shape)
print(new_grade)

# show Brad Exam2 score
print('Brad Exam2 score :', new_grade[1, 1])

# show Cassie Exam1, Exam2 score
print('Cassie score :', new_grade[2, :])

# show Exam1 score
print('Exam score :', new_grade[:, 0])
```
